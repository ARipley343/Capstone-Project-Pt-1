```python
#Loading in all key elements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import GridSearchCV
```

```python
#Loading in Dataframe
df = pd.read_csv('Data/df_dummy.csv')
df.head()
```

# Finding the Baseline

```python
# Using a dummyRegressor to find our baseline score. 
X = df.drop(columns = 'price')
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
#Set random state to 42 so it can be easily recreated

dummy_regr = DummyRegressor(strategy = 'mean')
dummy_regr.fit(X_train, y_train)
```

```python
y_pred_baseline = dummy_regr.predict(X_test)

baseline_mse = mean_squared_error(y_test, y_pred_baseline, squared = False)
print(baseline_mse)
```

> This gives us a baseline of 21247.25 for our RMSE score. For the models to be successful they must score LOWER than this number. 

# Creating the Models

> I am going to start with a basic Linear Regression model using all available features to give us a starting point.

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
train_preds = lr.predict(X_train)
test_preds = lr.predict(X_test)
```

```python
# Training RMSE Score
mean_squared_error(y_train, train_preds, squared = False)
```

```python
# Test RMSE Score
mean_squared_error(y_test, test_preds, squared = False)
```

```python
# Test R2 Score for refrence
r2_score(y_test, test_preds)
```

> For our first model we were able to beat the baseline RMSE score by a good margin! Now lets try a baseline for our other model 'RandomForestRegressor'

```python
rfr = RandomForestRegressor(random_state = 42)
rfr.fit(X_train, y_train)
```

```python
train_preds = rfr.predict(X_train)
test_preds = rfr.predict(X_test)
```

```python
# Training RMSE Score
mean_squared_error(y_train, train_preds, squared = False)
```

```python
# Test RMSE Score
mean_squared_error(y_test, test_preds, squared = False)
```

```python
# Test R2 Score for refrence
r2_score(y_test, test_preds)
```

> With our test score improving on the base Random Forest Regressor I am going to go ahead and create a param_grid for it to tune for the best paramaters.

```python
#Setting up the pramater grid with more options.
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]}
```

```python
rfr_cv = GridSearchCV(estimator = rfr, param_grid = param_grid, cv = 5, scoring = 'neg_root_mean_squared_error', n_jobs = -1)
```

```python
rfr_cv.fit(X_train, y_train)
```

```python
print("Top Parameters:", rfr_cv.best_params_)
```

> This gives us our best parameters and random forest regressor Let run it again specifically with these parameters to see what we get.

```python
rfr = RandomForestRegressor(random_state = 42, max_depth = 30, min_samples_leaf = 1, min_samples_split = 2, n_estimators = 300)
rfr.fit(X_train, y_train)
```

```python
train_preds = rfr.predict(X_train)
test_preds = rfr.predict(X_test)
```

```python
# Training RMSE Score
mean_squared_error(y_train, train_preds, squared = False)
```

```python
# Test RMSE Score
mean_squared_error(y_test, test_preds, squared = False)
```

```python
r2_score(y_test, test_preds)
```

> This gives us our best RMSE score at 10661.93 and our best R2 score of 0.75 Now lets try adjusting the features to see if we get any better results.

```python
plt.figure(figsize = (10, 18))
sns.heatmap(df.corr(numeric_only = True)[['price']].sort_values(by = 'price', ascending = False),
            vmin = -1,
            vmax = 1,
            cmap = 'coolwarm',
            annot = True);
```

> Looking back at the heat map for our target feature, I want to see if reducing the number of features in the model helps or hurts the RMSE score. So I will start by only using features with a corr value above 0.05. I dont want to remove too many features as I know from experience this tends to not help. My goal is to see if removing the least usefull features it can help the model tune a bit better. 

```python
# Creating new DF with only the desired columns 
threshold = 0.05
corr_mat = df.corr()
target_corr = corr_mat['price'].abs()
selected_columns = target_corr[(target_corr >= threshold) | (target_corr <= -threshold)].index # Make sure to include positive and negative correlations from this threshold. 
df_corr = df[selected_columns]
df_corr.head()
```

>With this new data frame I will rerun the RandomForestRegressor as before to see the results. 

```python
X = df_corr.drop(columns = 'price')
y = df_corr['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
#Set random state to 42 so it can be easily recreated
```

```python
rfr = RandomForestRegressor(random_state = 42, max_depth = 30, min_samples_leaf = 1, min_samples_split = 2, n_estimators = 300)
rfr.fit(X_train, y_train)
```

```python
train_preds = rfr.predict(X_train)
test_preds = rfr.predict(X_test)
```

```python
# Training RMSE Score
mean_squared_error(y_train, train_preds, squared = False)
```

```python
# Test RMSE Score
mean_squared_error(y_test, test_preds, squared = False)
```

```python
r2_score(y_test, test_preds)
```

> Unfortuantely this did not help. So our original tuned RandomForestRegressor will be our final model with and RMSE Score of 10661.93 and our best R2 score of 0.75. This is still a huge improvement on our baseline and an overall success.

My final thoughts on this analysis would be that while it was a success, I do believe more data could provide an even better model. Many of our features had values with very unique counts and I believe had we had 10,000 or more cars with more variablility in this data set it would have produced a better model.
